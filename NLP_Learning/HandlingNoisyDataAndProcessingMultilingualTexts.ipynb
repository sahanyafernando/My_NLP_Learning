## Part 1: Handling Noisy Datasets - Types, Emojis, Abbreviations

```python
# Load Dataset
import pandas as pd
df = pd.read_csv("noisy_dataset.csv")
print("Original Dataset")
print(df.head())
```

Removing Special Characters and NUmbers using Regex

```python
print("-Cleaning Text using Regex:")
import re

def clean_text(text):
  text = re.sub(r'[^\w\s]', '', text) # Remove special characters and punctuations
  text = re.sub(r'\d+', '', text) # Remove numbers
  return text.lower()

df['cleaned_text'] = df['text'].apply(clean_text)
print(df[["text", "cleaned_text"]].head())
```

Removing Emojis using emoji library

```python
print("- Removing Emojis")
import emoji

def remove_emojis(text):
  return emoji.replace_emoji(text, replace='')

df['No_Emojis'] = df['text'].apply(remove_emojis)
print(df[["text", "No_Emojis"]].head())
```

Expanding Abbreviations using a simple dictionary

```python
print("- Expanding Abbreviations")
abbr_dict = {"u": "you", "ur": "your", "lol": "laughing out loud", "brb": "be right back"}

def expand_abbreviations(text):
  words = text.split()
  return " ".join([abbr_dict[word] if word in abbr_dict else word for word in words])

df['expanded_text'] = df['text'].apply(expand_abbreviations)
print(df[["text", "expanded_text"]].head())
```

## Part 2: Processing Code-Mixed (Multilingual) Text

Load the code mixed dataset

```python
df = pd.read_csv("noisy_dataset.csv")
df.head()
```

### Detecting Language using langdetect

```python
# Detecting Language using langdetect
print("- Detecting Language")
from langdetect import detect

def detect_language(text):
  try:
    return detect(text)
  except:
    return "Unknown"

df["Detected_language"] = df["text"].apply(detect_language)
print(df[["text", "Detected_language"]])
```

```python
# Translating Non-English Text using Google Translator
print("- Translating Non-English Text")
from deep_translator import GoogleTranslator

def translate_text(text):
  try:
    lang = detect(text)
    if lang != "en":
           return GoogleTranslator(source=lang, target="en").translate(text)
    return text
  except:
    return text

df["Translated_text"] = df["text"].apply(translate_text)
print(df[["text", "Translated_text"]])

print("Processing Complete")
```
